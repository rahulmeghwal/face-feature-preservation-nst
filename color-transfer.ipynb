```python
from PIL import Image,ImageEnhance,ImageFilter
import matplotlib.pyplot as plt

def blur_style_image(file):
    image = Image.open(file)
    plt.imshow(image)
    plt.show()

    # ksize
    ksize = (15, 15)
    
    image = image.filter(ImageFilter.BoxBlur(8))
    image = image.filter(ImageFilter.BoxBlur(8))
    image = image.filter(ImageFilter.BoxBlur(8))

    plt.imshow(image)
    plt.show()

    result = ImageEnhance.Color(image)
    
    # showing resultant image
    result = result.enhance(1.2)
    splits = file.split(".")
    result.save(splits[0] + "-blurred." + splits[1])

    plt.imshow(result)
    plt.show()
```

```python
all_style_images = os.listdir(BASE_DIR + "/data/style-images/")
for style_image in all_style_images:
    blur_style_image(BASE_DIR + "/data/style-images/" + style_image)
```
